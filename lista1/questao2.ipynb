```python
from decimal import Decimal, getcontext, ROUND_HALF_UP

getcontext().prec = 3
getcontext().rounding = ROUND_HALF_UP

X_VAL = Decimal('0.5')
GRAU = 7
```

```python
def factorial_decimal(n):
    if n == 0 or n == 1:
        return Decimal('1')
    fat = Decimal('1')
    for i in range(2, n + 1):
        fat = fat * Decimal(str(i))
    return fat

def sen_polynomial_maclaurin(grau):
    coefficients = {}
    terms_str = []
    
    for n in range((grau + 1) // 2):
        potencia = 2 * n + 1
        sinal = Decimal('1') if n % 2 == 0 else Decimal('-1')
        denominador = factorial_decimal(potencia)
        
        coef = sinal / denominador 
        coefficients[potencia] = coef
        terms_str.append(f"({coef})*x^{potencia}")
    
    polynomial_str = "P(x) = " + " + ".join(terms_str).replace("+ -", "- ")
    return coefficients, polynomial_str

def solve_item_a(grau):
    coefs, pol_str = sen_polynomial_maclaurin(grau)
    print("a) Polinômio de Maclaurin gerado:")
    print(pol_str)
    return coefs

coefficients = solve_item_a(GRAU)
```

```python
def without_alinhamento(x, coefs):
    potencias = {1: x}
    x_atual = x
    for i in range(2, max(coefs.keys()) + 1):
        x_atual = x_atual * x
        potencias[i] = x_atual
        
    soma = Decimal('0')
    primeiro_termo = True
    
    for pot, coef in coefs.items():
        termo = coef * potencias[pot]
        if primeiro_termo:
            soma = termo
            primeiro_termo = False
        else:
            soma = soma + termo
            
    return soma

def solve_item_b(x_val, coefs):
    aprox = without_alinhamento(x_val, coefs)
    print(f"b) Aproximação sem aninhamento: {aprox}")
    return aprox

aprox_b = solve_item_b(X_VAL, coefficients)
```

```python
def polynomial_aninhada(coefs):
    potencias = sorted(coefs.keys(), reverse=True)
    
    expressao = f"{coefs[potencias[0]]}"
    for pot in potencias[1:]:
        expressao = f"({coefs[pot]} + x^2 * {expressao})"
        
    expressao = f"x * {expressao}"
    return expressao

def solve_item_c(coefs):
    expr_aninhada = polynomial_aninhada(coefs)
    print("c) Polinômio reescrito na forma aninhada:")
    print(f"P(x) = {expr_aninhada}")

# Execução do Item C
solve_item_c(coefficients)
```

```python
def aninhada(x, coefs):
    potencias = sorted(coefs.keys(), reverse=True)
    
    x_quad = x * x
    resultado = coefs[potencias[0]]
    
    for pot in potencias[1:]:
        resultado = resultado * x_quad
        resultado = resultado + coefs[pot]
        
    resultado = resultado * x
    return resultado

def solve_item_d(x_val, coefs):
    aprox = aninhada(x_val, coefs)
    print(f"d) Aproximação com aninhamento: {aprox}")
    return aprox

aprox_d = solve_item_d(X_VAL, coefficients)
```
