```python
# a)

T_A = 25.0  # Temperatura ambiente
T_0 = 90.0  # Temperatura inicial
K = 0.3     # Constante de resfriamento
T_ALVO = 40.0 # Temperatura alvo

def func_resfriamento(t_ambiente, t_inicial, k, t_alvo):

    def func(t):
        # f(t) = Ta + (T0 - Ta)*e^(-kt) - Talvo
        temperatura_t = t_ambiente + (t_inicial - t_ambiente) * math.exp(-k * t)
        return temperatura_t - t_alvo
    return func

def solve_item_a():
    f_t = func_resfriamento(T_A, T_0, K, T_ALVO)
    print("a) Função f(t) = 0 formulada e encapsulada com sucesso.")
    return f_t

# Execução do Item A
f_problema = solve_item_a()
```

```python
# b)

def metodo_bisseccao(f, a, b, tol=1e-5, max_iter=1000):
    # verifica pelo Teorema de Bolzano se a raiz está garantida no intervalo
    if f(a) * f(b) > 0:
        return None
    
    t_ant = a # inicializa o t anterior para o critério de parada
    
    for _ in range(max_iter):
        t_m = (a + b) / 2.0  # Ponto médio
        
        # criterio de parada: diferença entre as aproximações consecutivas
        if abs(t_m - t_ant) < tol:
            return t_m
            
        # atualiza os limites do intervalo
        if f(a) * f(t_m) < 0:
            b = t_m
        else:
            a = t_m
            
        t_ant = t_m  # atualiza para a próxima iteração
        
    return None
```

```python
# c)

def buscar_intervalos_e_resolver(f, a_global, b_global, C, tol=1e-5):
    t_inicio = a_global
    
    while t_inicio < b_global:
        t_fim = min(t_inicio + C, b_global)
        
        if f(t_inicio) * f(t_fim) <= 0:
            print(f"c) Raiz detectada no subintervalo: [{t_inicio}, {t_fim}]")
            raiz = metodo_bisseccao(f, t_inicio, t_fim, tol)
            return raiz
            
        t_inicio += C
        
    return None
```

```python
# d)

def solve_item_d(f):
    # Parâmetros definidos no enunciado da letra D
    A_GLOBAL = 0.0
    B_GLOBAL = 30.0
    COMPRIMENTO_C = 2.0
    TOLERANCIA = 1e-5
    
    print("d) Iniciando a varredura e o cálculo...\n")
    tempo_raiz = buscar_intervalos_e_resolver(
        f, 
        A_GLOBAL, 
        B_GLOBAL, 
        COMPRIMENTO_C, 
        TOLERANCIA
    )
    
    print("\n=== RESULTADO FINAL ===")
    if tempo_raiz is not None:
        print(f"O dispositivo atinge 40°C no instante t ≈ {tempo_raiz:.4f} minutos.")
        
        # Prova real
        temperatura_final = f(tempo_raiz) + T_ALVO # T(t) = f(t) + 40
        print(f"Prova real (Temperatura neste instante): {temperatura_final:.4f}°C")
    else:
        print("Nenhuma raiz foi encontrada no intervalo fornecido.")

# Execução do Item D passando a função gerada na letra A
solve_item_d(f_problema)
```
