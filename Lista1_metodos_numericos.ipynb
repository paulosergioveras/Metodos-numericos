```python
import matplotlib.pyplot as plt
import math
```

```python
def plot_function(f, a, b, n):

    # divide o intervalo em partes iguais
    step = (b - a) / (n - 1)

    x_points = []
    y_points = []

    for i in range(n):
        x = a + i * step # gera os pontos
        y = f(x)
        #armazena os valores
        x_points.append(x)
        y_points.append(y)

    # plotando o gráfico
    plt.figure(figsize=(10, 6))
    plt.plot(x_points, y_points, 'b-', linewidth=2, label=f'y = f(x)') #conecta os pontos com segmentos de reta
    plt.scatter(x_points, y_points, color='red', s=20, zorder=5, label='Pontos amostrados') #mostra os pontos amostrados
    plt.grid(True, alpha=0.3)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f' f no intervalo [{a}, {b}] com {n} pontos')
    plt.legend()
    plt.show()

    print(f"Dominio: [{a}, {b}]")
    print(f"Numero de pontos: {n}")
    print(f"Tamanho do passo: {step:.6f}")
    print(f"Intervalo de x: de {x_points[0]:.6f} a {x_points[-1]:.6f}")

```

```python
# Exemplo de gráfico com a função quadrática f(x) = x² - 4x + 3
def quadratic_function(x):
    return x**2 - 4*x + 3
```

```python

plot_function(quadratic_function, -2, 6, 50)
```

```python
# Exemplo de gráfico da função trigonompetrica f(x) = sin(3x) * cos(x)
def trig_function(x):
    return math.sin(3*x) * math.cos(x)
```

```python
plot_function(trig_function, 0, 2*math.pi, 80)
```

Questão 5:
a) \
    Dada a equação x² + x − 1 = 0 \
    g: função de iteração \
    x0: chute inicial \
    tol: tolerância para o critério de parada \
    max_iter: máximo de iterações


```python
def fixed_point(g, x0, tol=1e-6, max_iter=100):

    x = x0

    for i in range(max_iter):
        x_next = g(x)


        if abs(x_next - x) < tol:
            print(f"Convergiu em {i+1} iterações")
            return x_next

        x = x_next

    print("Não converge")
    return x
```

```python
# Função de iteração:
def g1(x):
    return 1 / (1 + x)

def g2(x):
    return 1 - (1 / x)
```

```python
print("Para g1: ")
raiz1 = fixed_point(g1, x0=0.5)
print("Raiz para g1:", raiz1)

print("Para g2: ")
raiz2 = fixed_point(g2, x0=0.5)
print("Raiz para g2:", raiz2)
```

Questão 5: b) \
 g: função que define a iteração \
x0: chute inicial \
n_max: número máximo de iterações \
tol : tolerancia para o criterio de parada

```python
def fixed_point_iteration(g, x0, tol, n_max):
    x_prev = x0

    for i in range(n_max):
        try:
            x_next = g(x_prev)  #aplica a função ao valor atual para obter o próximo

            if abs(x_next - x_prev) < tol: #com base no criterio de parada informado no enunciado
                return x_next

            x_prev = x_next

        except (OverflowError, ValueError):
            return None

    return None
```

```python
import math
def g1(x):
    return math.cos(x)

x0 = 0
tol = 1e-8
n_max = 50

resultado = fixed_point_iteration(g1, x0, tol, n_max)

if resultado is not None:
    print(f"Para g(x) = cos (x) a aproximacao de ponto fixo é: {resultado:.10f}\n")
else:
    print(f"Para g(x) = cos (x) não converge mesmo após {n_max} iteracoes\n")


def g2(x):
    return math.exp(x) - 2

x1 = 2.0
tol1 = 1e-8
n_max1 = 50

resultado1 = fixed_point_iteration(g2, x1, tol1, n_max1)

if resultado1 is not None:
    print(f"Para g(x) = e^x - 2 a aproximacao de ponto fixo é: {resultado1:.10f}\n")
else:
    print(f"Para g(x) = e^x - 2 não converge mesmo após {n_max1} iteracoes\n")


```

Questão 5: d) Para que g($x$) convirja no intervalo [0,1] , ela precisa satisfazer duas condições:


* ∀ $x$ ∈ [0,1], o resultado g($x$) também deve estar nesse intervalo;
*  | g'($x$) | <= k < 1 , ∀ $x$ ∈ [0,1]  

\
Para $g_1(x) = \frac{1}{1+x}$ :  $g_1'(x) = -\frac{1}{(1+x)^2}$  \
Se $x = 0 \implies |g_1'(0)| = 1$ \
Se $x = 1 \implies |g_1'(1)| = 0,25$

*   No intervalo aberto $(0, 1]$, temos sempre | $g_1'($x$)$ | < 1
*   Para qualquer $x$ entre 0 e 1, $g_1(x)$ resulta em um valor entre $0,5$ e $1$ --> permanece dentro do intervalo, por isso $g_1'$ converge

\
Para $g_2(x) = 1 - \frac{1}{x}$ : $g_2'(x) = \frac{1}{x^2}$ \
Se $x = 0$ --> $g_2(x)$ falha pois teremos uma divisão por $0$

*   Para qualquer $x \in (0, 1]$, temos $|g_2'(x)| \geq 1$
*   Logo, $g_2(x)$ afasta os pontos da iteração de ponto fixo e diverge

\
Logo, $g_1(x)$ é a formulação mais adequada pois satisfaz o teorema de ponto fixo ($|g'(x)| < 1$) fazendo com que o erro diminua a cada iteração.

Questão 6:  a)  e  b) \
$f(x)$ = $x³ - x - 2 = 0$ \
$f'(x)$ = $x³ - x - 2 = 0$ \
$f (x) = 0$ \
x0 = 1.5  \
tol = 1e-6

```python
def newton(f, df, x0, tol, n_max):
    x_prev = x0

    for i in range(n_max):
        # caso a derivada seja 0, evita essa divisao
        derivada = df(x_prev)
        if derivada == 0:
            print("Derivada = 0")
            return None, i

        x_next = x_prev - f(x_prev) / derivada

        if abs(x_next - x_prev) < tol:
            return x_next, i + 1

        x_prev = x_next

    return None
```

```python
def f(x):
    return x**3 - x - 2

def df(x):
    return 3*x**2 - 1

x0 = 1.5
tol = 1e-6
n_max = 50

resultado, i = newton(f, df, x0, tol, n_max)

if resultado:
    print(f"Raiz encontrada usando o método de Newton: {resultado:.5f}")
    print(f"Número de iterações: {i}")
else:
    print("O método não convergiu.")
```

Questão 6: c)\
Intervalo $[1,2]$ \
Tolerância: $1e-6$

```python
def bisection(f, a, b, tol, n_max):
    if f(a) * f(b) >= 0:
        print("O método da bisseção falhou: f(a) e f(b) devem ter sinais opostos.")
        return None, 0

    x_prev = a
    for i in range(n_max):
        x_next = (a + b) / 2

        if abs(x_next - x_prev) < tol:
            return x_next, i + 1

        if f(a) * f(x_next) < 0:
            b = x_next
        else:
            a = x_next

        x_prev = x_next

    return None
```

```python
def f(x):
    return x**3 - x - 2

a = 1.0
b = 2.0
tol = 1e-6
max_iter = 100

resultado, i = bisection(f, a, b, tol, max_iter)

if resultado is not None:
    print(f"Raiz encontrada usando o método da Bisseção: {resultado:.5f}")
    print(f"Número de iterações: {i}")
else:
    print("O método da bisseção não convergiu.")
```
